# Ejemplos de conjuntos convexos

```python
# Importing necessary libraries 
import numpy as np
import matplotlib.pyplot as plt
```

**Semiespacio**

   El conjunto $ \mathcal{S} = \{(x_1, x_2, x_3) \,/\, x_1 + 2x_2 - x_3 \leq 4\} $ representa un semiespacio en tres dimensiones. Un semiespacio es una división del espacio euclidiano que incluye todos los puntos de un lado de un plano determinado por la desigualdad dada.

```python
# Parameters for the 3D plots
grid_size_3d = 0.5
max_val_3d = 8
x_3d = np.arange(-max_val_3d, max_val_3d, grid_size_3d)
y_3d = np.arange(-max_val_3d, max_val_3d, grid_size_3d)
x_3d, y_3d = np.meshgrid(x_3d, y_3d)

# First semispace
z1_3d = 4 - x_3d - 2*y_3d  

# Create four separate figures for each example
# Figure for the semispace x + 2y - z <= 4
fig_semispace = plt.figure(figsize=(6, 6))
ax_semispace = fig_semispace.add_subplot(111, projection='3d')
ax_semispace.plot_surface(x_3d, y_3d, z1_3d, alpha=0.5, rstride=100, cstride=100, color='r', edgecolor='none')
ax_semispace.set_title('Semispace x1 + 2x2 - x3 ≤ 4$')
plt.show()
```

**Intersección de Dos Semiespacios**

   El conjunto $ \mathcal{S} = \{(x_1, x_2, x_3) \,/\, x_1 + 2x_2 - x_3 \leq 4, \, 2x_1 - x_2 + x_3 \leq 6\} $ es la intersección de dos semiespacios en tres dimensiones. La intersección es convexa y representa el conjunto de puntos que satisfacen ambas desigualdades simultáneamente.

```python
z2_3d = (6 - 2*x_3d + y_3d) / 1  # Second semispace

# Figure for the intersection of two semispaces
fig_intersection = plt.figure(figsize=(6, 6))
ax_intersection = fig_intersection.add_subplot(111, projection='3d')
ax_intersection.plot_surface(x_3d, y_3d, np.minimum(z1_3d, z2_3d), alpha=0.5, rstride=100, cstride=100, color='g', edgecolor='none')
ax_intersection.set_title('Intersection of two semispaces')
plt.show()
```

**Cono**

El conjunto $ \mathcal{S} = \{(x_1, x_2) \,/\, x_2 \geq |x_1|\} $ forma un cono en dos dimensiones. Este conjunto incluye todos los puntos en el plano donde el valor de $ x_2 $ es mayor o igual al valor absoluto de $ x_1 $, formando una región en forma de V que se extiende infinitamente hacia arriba.

```python
# Parameters for the 2D cone plot
cone_grid_size = 0.1
cone_max_val = 5
x_cone = np.arange(-cone_max_val, cone_max_val, cone_grid_size)
y_cone = np.abs(x_cone)

# Figure for the cone x2 >= |x1|
fig_cone_2d = plt.figure(figsize=(6, 6))
ax_cone_2d = fig_cone_2d.add_subplot(111)
ax_cone_2d.fill_between(x_cone, y_cone, cone_max_val, color='b', alpha=0.5)
ax_cone_2d.set_xlim(-cone_max_val, cone_max_val)
ax_cone_2d.set_ylim(0, cone_max_val)
ax_cone_2d.set_title('Cone x2 ≥ |x1|')
plt.show()
```

**Círculo**

El conjunto $ \mathcal{S} = \{(x_1, x_2) \,/\, x_1^2 + x_2^2 \leq 4\} $ es el área dentro de un círculo centrado en el origen con un radio de 2 unidades en el plano $ x_1-x_2 $. Este conjunto es también convexo y contiene todos los puntos cuya distancia al origen es menor o igual a 2.

```python
# Parameters for the 2D circle plot
circle_max_val = 4
x_circle = np.linspace(-circle_max_val, circle_max_val, 400)
y_circle = np.linspace(-circle_max_val, circle_max_val, 400)
x_circle, y_circle = np.meshgrid(x_circle, y_circle)
z_circle = x_circle**2 + y_circle**2

# Figure for the filled circle x1^2 + x2^2 <= 4
fig_circle_filled = plt.figure(figsize=(6, 6))
ax_circle_filled = fig_circle_filled.add_subplot(111)
ax_circle_filled.contourf(x_circle, y_circle, z_circle, levels=[0, 4], colors='m', alpha=0.5)
ax_circle_filled.set_xlim(-circle_max_val, circle_max_val)
ax_circle_filled.set_ylim(-circle_max_val, circle_max_val)
ax_circle_filled.set_aspect('equal', 'box')
ax_circle_filled.set_title('Circle x1^2 + x2^2 ≤ 4')
plt.show()
```
