# Funciones Convexas

```python
# Importing necessary libraries 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the functions for 2D plots
def f_linear(x):
    return 3*x - 4

def f_absolute(x):
    return np.abs(x)

def f_quadratic(x):
    return x**2 - 2*x

def f_sqrt(x):
    # Function is defined for non-negative x
    return 3 - np.sqrt(x)

# Define the function for 3D plots
def f_3d(x1, x2):
    return 2*x1**2 + x2**2 - 2*x1*x2

def f_3d_2(x1, x2, x3):
    return x1**4 + 2*x2**2 + 3*x3**2 - 4*x1 - 4*x2*x3

# Create a range of x values for plotting 2D functions
x_values = np.linspace(-10, 10, 200)
x_values_non_negative = x_values[x_values >= 0]  # For the square root function which is only defined for x >= 0

# Create a grid for the 3D function f_3d
x1_values = np.linspace(-3, 3, 100)
x2_values = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1_values, x2_values)
Z = f_3d(X1, X2)

# Create a grid for the second 3D function f_3d_2
x3_values = np.linspace(-3, 3, 100)
X1_2, X2_2, X3_2 = np.meshgrid(x1_values, x2_values, x3_values)
Z_2 = f_3d_2(X1_2, X2_2, X3_2)

# Plot a slice for a fixed value of x3 for the function f_3d_2
x3_fixed = 0  # We can change this value to see different slices of the function
Z_2_slice = f_3d_2(X1, X2, x3_fixed)
```

1. **Función Lineal**

   $$ f(x) = 3x - 4 $$
   
    Esta es una función lineal con una pendiente de 3 y una intersección en y de -4. Es una función continua y recta en toda su dominio.

```python
# Plot the 2D functions separately
# f(x) = 3x - 4
plt.figure(figsize=(6, 4))
plt.plot(x_values, f_linear(x_values), label="f(x) = 3x - 4")
plt.title("Linear Function")
plt.legend()
plt.show()
```

2. **Función de Valor Absoluto**

   $$ f(x) = |x| $$
   
   La función de valor absoluto crea una forma de V, siendo siempre positiva o cero. Refleja los valores negativos de x en el eje positivo de y.

```python
# f(x) = |x|
plt.figure(figsize=(6, 4))
plt.plot(x_values, f_absolute(x_values), label="f(x) = |x|")
plt.title("Absolute Value Function")
plt.legend()
plt.show()
```

3. **Función Cuadrática**

   $$ f(x) = x^2 - 2x $$
   
   Una parábola con su vértice desplazado desde el origen. Esta función es cóncava hacia arriba, con un mínimo local que puede ser encontrado mediante el cálculo.

```python
# f(x) = x^2 - 2x
plt.figure(figsize=(6, 4))
plt.plot(x_values, f_quadratic(x_values), label="f(x) = x^2 - 2x")
plt.title("Quadratic Function")
plt.legend()
plt.show()
```

4. **Función Raíz Cuadrada**

   $$ f(x) = 3 - x^{1/2} $$ para $$ x \geq 0 $$
   
   Esta función está definida solo para valores no negativos de x. Se comporta de manera decreciente y se aplana a medida que x aumenta.

```python
# f(x) = 3 - x^(1/2), x >= 0
plt.figure(figsize=(6, 4))
plt.plot(x_values_non_negative, f_sqrt(x_values_non_negative), label="f(x) = 3 - x^(1/2), x >= 0")
plt.title("Square Root Function")
plt.legend()
plt.show()
```

5. **Función en 3D**

   $$ f(x_1, x_2) = 2x_1^2 + x_2^2 - 2x_1x_2 $$
   
   Es una función de dos variables que produce una superficie en 3D. Esta superficie puede representar distintas formas geométricas dependiendo de los coeficientes y términos presentes.

```python
# Plot the 3D function f_3d
fig_3d = plt.figure(figsize=(8, 6))
ax_3d = fig_3d.add_subplot(111, projection='3d')
ax_3d.plot_surface(X1, X2, Z, cmap='viridis')
ax_3d.set_title("3D Function f(x1, x2) = 2x1^2 + x2^2 - 2x1x2")
plt.show()
```

6. **Función en 4D (Representada como una Sección Transversal en 3D)**

   $$ f(x_1, x_2, x_3) = x_1^4 + 2x_2^2 + 3x_3^2 - 4x_1 - 4x_2x_3 $$
   
   Dado que es difícil visualizar cuatro dimensiones, esta función se representa en tres dimensiones fijando $ x_3 $ a un valor constante.

```python
# Plot the slice of the second 3D function f_3d_2
fig_3d_2 = plt.figure(figsize=(8, 6))
ax_3d_2 = fig_3d_2.add_subplot(111, projection='3d')
ax_3d_2.plot_surface(X1, X2, Z_2_slice, cmap='viridis')
ax_3d_2.set_title("3D Function Slice f(x1, x2, x3) for x3=0")
plt.show()
```
