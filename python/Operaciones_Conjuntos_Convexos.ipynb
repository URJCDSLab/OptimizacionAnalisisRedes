# Exploración Interactiva de Operaciones con Conjuntos Convexos

Este notebook ofrece una exploración detallada y visual de las propiedades fundamentales de los conjuntos convexos y cómo se comportan bajo diversas operaciones matemáticas. Utilizaremos Python para visualizar y analizar estos conceptos esenciales.

## Comprensión Profunda de los Conjuntos Convexos

### ¿Qué es un Conjunto Convexo?

Un conjunto convexo en un espacio vectorial es un conjunto $ C $ que satisface la siguiente propiedad geométrica: para cualquier par de puntos $ x, y \in C $, el segmento de línea que los une, dado por $ \lambda x + (1 - \lambda) y $ para $ 0 \leq \lambda \leq 1 $, está completamente contenido en $ C $. Esta definición matemática asegura que el conjunto no presenta "recovecos" o "agujeros".

### Características Esenciales de los Conjuntos Convexos

Los conjuntos convexos son fundamentales en diversas áreas, como el análisis matemático y la optimización. La convexidad asegura que no existen "curvas hacia adentro" en el borde del conjunto. Matemáticamente, una función $ f: C \rightarrow \mathbb{R} $ es convexa si $ C $ es un conjunto convexo y se cumple que:

$$ f(\lambda x + (1 - \lambda) y) \leq \lambda f(x) + (1 - \lambda) f(y) \, \text{para todo} \, x, y \in C \, \text{y} \, 0 \leq \lambda \leq 1 $$

## Teoría Detrás de las Operaciones con Conjuntos Convexos

1. **Intersección de Conjuntos Convexos:**
   - Dados dos conjuntos convexos $ A $ y $ B $, su intersección $ A \cap B $ es convexa. Esto se debe a que cualquier segmento de línea dentro de $ A \cap B $ debe estar en ambos $ A $ y $ B $, y por ende, en su intersección.

2. **Suma de Conjuntos Convexos:**
   - La suma de dos conjuntos convexos $ A $ y $ B $, denotada por $ A + B $, que contiene elementos de la forma $ a + b $ con $ a \in A $ y $ b \in B $, también forma un conjunto convexo.

3. **Resta de Conjuntos Convexos:**
   - Similar a la suma, la resta de conjuntos convexos, definida como $ A - B = \{ a - b \,|\, a \in A, b \in B \} $, resulta en un conjunto convexo.

## Objetivos de Visualización

En este notebook, nos enfocaremos en ilustrar visualmente estas propiedades y operaciones. Demostraremos cómo la intersección, suma y resta afectan la forma y la naturaleza de los conjuntos convexos, proporcionando una comprensión más profunda de su teoría y aplicaciones prácticas.

# Intersección de Conjuntos Convexos

Este gráfico ilustra la intersección de dos conjuntos convexos, representados por círculos, para demostrar visualmente cómo se determina la intersección en un contexto geométrico.

## Configuración de los Conjuntos Convexos

Los conjuntos convexos en esta visualización son representados por dos círculos:

- El **Conjunto A** está centrado en el punto (0, 0) y tiene un radio de 1.
- El **Conjunto B** está centrado en el punto (0.5, 0) y también tiene un radio de 1.

Dado que ambos círculos tienen un radio de 1 y sus centros están separados por una distancia de 0.5, se superponen parcialmente. Esta superposición crea una región de intersección.

## Visualización de la Intersección

En el gráfico:

- El **Conjunto A** está rellenado con color azul.
- El **Conjunto B** está rellenado con color rojo.
- La **área de intersección** de los conjuntos se visualiza donde los colores azul y rojo se superponen, resultando en una mezcla de estos colores.

Esta representación visual permite identificar fácilmente la intersección entre los dos conjuntos convexos y proporciona una comprensión intuitiva de cómo se determinan las áreas compartidas en geometrías simples como los círculos.


```python
import numpy as np
import matplotlib.pyplot as plt
# Define two convex sets (circles) with adjusted positions for intersection
theta = np.linspace(0, 2 * np.pi, 100)
x1, y1 = np.cos(theta), np.sin(theta)  # First circle centered at (0, 0)
x2, y2 = np.cos(theta) + 0.5, np.sin(theta)  # Second circle centered at (0.5, 0)

# Plot the two sets and their overlap
plt.figure(figsize=(8, 6))
plt.fill(x1, y1, alpha=0.5, color='blue', label='Conjunto A')
plt.fill(x2, y2, alpha=0.5, color='red', label='Conjunto B')
plt.title('Intersección de Conjuntos Convexos A y B')
plt.legend()
plt.grid(True)
plt.show()
```

# Suma y Resta de Conjuntos Convexos Muestreados

Este análisis muestra la suma y la resta de dos conjuntos convexos, representados por círculos, utilizando un enfoque de muestreo. Se seleccionaron puntos de cada círculo y luego se calcularon la suma y la resta de estos puntos muestreados para visualizar los resultados.

## Proceso de Muestreo y Cálculo

Se muestrearon 100 puntos de dos círculos:

- El **Primer Círculo** centrado en (0, 0) con radio 1.
- El **Segundo Círculo** centrado en (0.5, 0) con radio 1.

Se realizó la suma y la resta de todos los pares de puntos muestreados de los dos círculos, resultando en dos conjuntos de puntos:

- **Suma de Conjuntos**: Se sumó cada punto del primer círculo con cada punto del segundo círculo.
- **Resta de Conjuntos**: Se restó cada punto del segundo círculo de cada punto del primer círculo.

## Visualización de Resultados

El gráfico muestra dos paneles:

1. **Suma de Conjuntos Convexos (Panel Izquierdo)**: 
   - Cada punto verde representa el resultado de la suma de un punto del primer círculo con un punto del segundo círculo.
   - La región cubierta por estos puntos muestra cómo se extiende la suma de los dos conjuntos más allá de los límites individuales de cada círculo.

2. **Resta de Conjuntos Convexos (Panel Derecho)**:
   - Cada punto morado representa el resultado de la resta de un punto del segundo círculo de un punto del primer círculo.
   - La dispersión de estos puntos refleja las diferencias vectoriales entre los puntos de los dos conjuntos.

Estas visualizaciones proporcionan una comprensión aproximada de las operaciones de suma y resta en conjuntos convexos mediante un enfoque de muestreo. Si bien estas representaciones son aproximadas, basadas en un número finito de puntos muestreados, ofrecen una perspectiva intuitiva de las regiones que podrían cubrir la suma y la resta de dos conjuntos convexos.

```python
# Definir una función para muestrear puntos de un círculo
def sample_circle(center, radius, num_samples=100):
    angles = np.linspace(0, 2 * np.pi, num_samples)
    x = center[0] + radius * np.cos(angles)
    y = center[1] + radius * np.sin(angles)
    return x, y

# Muestrear puntos de los dos círculos
x1, y1 = sample_circle(center=(0, 0), radius=1)
x2, y2 = sample_circle(center=(0.5, 0), radius=1)

# Calcular la suma y la resta de los conjuntos muestreados
sum_x, sum_y = x1[:, None] + x2[None, :], y1[:, None] + y2[None, :]
diff_x, diff_y = x1[:, None] - x2[None, :], y1[:, None] - y2[None, :]

# Aplanar los arrays para la visualización
sum_x, sum_y = sum_x.flatten(), sum_y.flatten()
diff_x, diff_y = diff_x.flatten(), diff_y.flatten()

# Graficar la suma y la resta de los conjuntos
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Graficar la suma
axes[0].scatter(sum_x, sum_y, color='green', s=10)
axes[0].set_title('Suma de Conjuntos Convexos')
axes[0].grid(True)
axes[0].set_aspect('equal', adjustable='box')

# Graficar la resta
axes[1].scatter(diff_x, diff_y, color='purple', s=10)
axes[1].set_title('Resta de Conjuntos Convexos')
axes[1].grid(True)
axes[1].set_aspect('equal', adjustable='box')

plt.show()
```
