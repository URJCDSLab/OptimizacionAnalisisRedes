## Teoría Básica de Epígrafos

### Definición General de Epígrafo

El epígrafo de una función es un concepto clave en análisis matemático, especialmente útil en áreas como la optimización y la teoría de la convexidad. De manera general, el epígrafo de una función $ f $ es el conjunto de todos los puntos que se encuentran por encima o en el gráfico de $ f $. Esto se traduce en la consideración de todos los puntos en un espacio que no están "por debajo" de la función. Formalmente, para una función $ f: \mathbb{R} \rightarrow \mathbb{R} $, el epígrafo se define como:

$$ \text{Epigrafo}(f) = \{ (x, y) \in \mathbb{R}^2 \,|\, y \geq f(x) \} $$

Esta definición implica que para cada punto $ (x, y) $ en el epígrafo, $ y $ es al menos igual al valor de $ f(x) $, representando así todas las posiciones "sobre" o "en" la función.

### Aplicación en la Convexidad

El estudio de los epígrafos es especialmente relevante en el análisis de la convexidad de funciones. Una función es convexa si su epígrafo también lo es, lo que significa que el segmento de línea que une cualquier par de puntos dentro del epígrafo también debe estar completamente contenido en el epígrafo.

## Ejemplo 1: Visualización del Epígrafo de una Función Cuadrática

### Función Cuadrática: $ f(x) = x^2 $

**Definición Específica del Epígrafo:** 
Para la función cuadrática $ f(x) = x^2 $, el epígrafo se define como el conjunto de todos los pares de coordenadas $ (x, y) $ que satisfacen la condición $ y \geq x^2 $. Esto incluye cada punto en el plano $ \mathbb{R}^2 $ que se encuentra en o por encima de la curva de la función cuadrática.

**Objetivo de la Visualización:**
La visualización que sigue ilustrará el epígrafo de la función cuadrática. Se usará un área sombreada en el gráfico para representar el epígrafo, brindando una interpretación visual clara de este conjunto y su significado matemático.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function and range for the quadratic function
def f(x):
    return x ** 2

x = np.linspace(-10, 10, 400)
y = f(x)

# Create the plot for the quadratic function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^2')
plt.fill_between(x, y, max(y), alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Epígrafo de la función f(x) = x^2')
plt.legend()
plt.grid(True)
plt.show()      
```

## Ejemplo 2: Epígrafo de una Función Definida a Trozos

### Función Definida a Trozos

Consideremos una función definida a trozos:

- $ f(x) = 0 $ para $ x < 2 $
- $ f(x) = 3 $ para $ x \geq 2 $

**Definición Específica del Epígrafo:** 
El epígrafo de esta función se define de la siguiente manera:

$$
\text{Epigrafo}(f) = \left\{ 
\begin{array}{ll}
(x, y) \in \mathbb{R}^2 : & x < 2 \, \text{y} \, y \geq 0, \\
(x, y) \in \mathbb{R}^2 : & x \geq 2 \, \text{y} \, y \geq 3.
\end{array}
\right.
$$

Esto significa que el epígrafo de esta función es el conjunto de todos los puntos $ (x, y) $ en el plano tal que:

- Para $ x < 2 $, cualquier punto donde $ y $ es mayor o igual a 0.
- Para $ x \geq 2 $, cualquier punto donde $ y $ es mayor o igual a 3.

**Objetivo de la Visualización:**
La visualización destacará cómo se representa el epígrafo de una función definida a trozos, mostrando claramente las áreas sombreadas que corresponden al conjunto definido por la función en diferentes intervalos de $ x $.

```python
# Define the piecewise function with a discontinuity
def f_piecewise(x):
    return np.where(x < 2, 0, 3)

x_piecewise = np.linspace(-10, 10, 1000)
y_piecewise = f_piecewise(x_piecewise)

# Create the plot for the piecewise function
plt.figure(figsize=(8, 6))
plt.plot(x_piecewise, y_piecewise, label='Función a trozos')
plt.fill_between(x_piecewise, y_piecewise, max(y_piecewise) + 1, alpha=0.3)
plt.scatter([2], [0], color='blue')
plt.scatter([2], [3], color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Epígrafo de una función definida a trozos con discontinuidad')
plt.legend()
plt.grid(True)
plt.show()
```
