# Autograd y Sympy


### Funciones univariantes

Supongamos que tenemos la función $f(x)=5x+10$ (una recta), y queremos evaluar $\nabla f(x_0=1)$

```python
def f(x):
    return (x[0]-3)**2+()

gradiente = grad(f)

x0 = 3.0

print('El gradiente en x0 es:', gradiente(x0))
```

```python
def f(x):
    return 5*x+10

gradiente = grad(f)

x0 = 1.0

print('El gradiente en x0 es:', gradiente(x0))
```

```python
# Definir la variable simbólica
x = sp.symbols('x')

# Definir la función de manera simbólica
f = 5*x + 10

# Calcular el gradiente (derivada) de la función
gradiente = sp.diff(f, x)

# Evaluar el gradiente en x0
x0 = 1.0
gradiente_en_x0 = gradiente.subs(x, x0)

print('El gradiente en x0 es:', gradiente_en_x0)
```

### Funciones multivariantes

Supongamos que tenemos definimos el plano $f(x_1,x_2)=5x_1+10x_2+4$, y queremos evaluar en $\mathbf{x}_0=[1,2]^{\top}$:

- $\nabla f(\mathbf{x}_0)$
- $\mathbf{H}_f(\mathbf{x}_0)$

```python
def plano(x): 
    return 5*x[0] + 10*x[1] + 4

x0 = np.array([1,2],dtype=float)

gradiente = grad(plano)
hessiana = hessian(plano)

print('El gradiente en x0 es:', gradiente(x0))
print('La hessiana en x0 es:\n', hessiana(x0))
```

```python
# Definir las variables simbólicas
x1, x2 = sp.symbols('x1 x2')

# Definir la función de manera simbólica
def plano_sympy(x1, x2):
    return 5*x1 + 10*x2 + 4

# Calcular el gradiente de la función
gradiente_sympy = sp.Matrix([sp.diff(plano_sympy(x1, x2), var) for var in (x1, x2)])

# Calcular la matriz Hessiana
hessiana_sympy = sp.hessian(plano_sympy(x1, x2), (x1, x2))

# Evaluar el gradiente y el hessiana en x0
x0 = np.array([1, 2], dtype=float)
gradiente_en_x0 = gradiente_sympy.subs({x1: x0[0], x2: x0[1]})
hessiana_en_x0 = hessiana_sympy.subs({x1: x0[0], x2: x0[1]})

print('El gradiente en x0 es:', gradiente_en_x0)
print('La hessiana en x0 es:\n', hessiana_en_x0)
```

### Funciones paramétricas

Supongamos que tenemos definimos el plano $f(x_1,x_2)=ax_1+bx_2+4$, y queremos evaluar en $\mathbf{x}_0=[1,2]^{\top}$:

- $\nabla f(\mathbf{x}_0)$
- $\mathbf{H}_f(\mathbf{x}_0)$

```python
def plano_con_parametros(x,a,b):
    return a*x[0] + b*x[1] + 4

a = 5 
b = 10

x0 = np.array([1,2],dtype=float)

gradiente = grad(plano_con_parametros)
hessiana = hessian(plano_con_parametros)

print('El gradiente en x0 es:', gradiente(x0,a,b))
print('La hessiana en x0 es:\n', hessiana(x0,a,b))
```

```python
# Definir las variables simbólicas
x1, x2, a, b = sp.symbols('x1 x2 a b')

# Definir la función de manera simbólica, incluyendo los parámetros a y b
def plano_con_parametros_sympy(x1, x2, a, b):
    return a*x1 + b*x2 + 4

# Calcular el gradiente de la función
gradiente_sympy = sp.Matrix([sp.diff(plano_con_parametros_sympy(x1, x2, a, b), var) for var in (x1, x2)])

# Calcular la matriz Hessiana
hessiana_sympy = sp.hessian(plano_con_parametros_sympy(x1, x2, a, b), (x1, x2))

# Valores de a, b y x0
a_val = 5
b_val = 10
x0 = np.array([1, 2], dtype=float)

# Evaluar el gradiente y el hessiana en x0, a y b
gradiente_en_x0 = gradiente_sympy.subs({x1: x0[0], x2: x0[1], a: a_val, b: b_val})
hessiana_en_x0 = hessiana_sympy.subs({x1: x0[0], x2: x0[1], a: a_val, b: b_val})

print('El gradiente en x0 es:', gradiente_en_x0)
print('La hessiana en x0 es:\n', hessiana_en_x0)
```
